<b>Walmart Data</b>
<p>Raw data from kaggle is obtained in 3 datasets: walmart_store, walmart_data and walmart_features</p>
<p>These data are merged into 1 dataset w_data</p>
<p>Integration of dataData is combined into 1 dataframe</p>

```python
#print(walmart_store, walmart_data, walmart_features)
w_store = np.array([[0, 0]] + np.array(walmart_store[["Type", "Size"]]).tolist()) #Store data converted into a lookup table
result = pd.merge(walmart_data, walmart_features, how='left', on=['Store', 'Date']) #Merge data and features data

print(['Store Floor Area', 'Dept', 'Date', 'Weekly_Sales', 'IsHoliday', 'Temperature', 'Fuel_Price', 'MarkDown', 'CPI', 'Unemployment'])
w_data = np.array(result)

new = []
for i in w_data:
    i[0] = w_store[i[0]][1] #Replace store number with floor area.
    i[7] = i[7] + i[8] + i[9] + i[10] + i[11] #Sum up markdown data
    i[8] = i[12]
    i[9] = i[13]
    new.append(i[0:9])

w_data = np.array(new)
print(w_data[10000])
```
